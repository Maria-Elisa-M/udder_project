```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
dirpath = os.getcwd()
data_path = os.path.join(dirpath, "..", "delpro_vms", "data")
# read vistit data
all_df = pd.read_csv(os.path.join(data_path, "delpro_visit_lactation.csv"))
```

```python
# find columns with teats
columns_q = [col for col in all_df.columns if col.split("_")[-1] in ["lf", "lr", "rr", "rf"]]
col_dict = {}
for col in  columns_q:
    key = "_".join(col.split("_")[0:-1])
    if key in col_dict.keys():
        col_dict[key].append(col)
    else:
        col_dict[key] = [col]
```

```python
for i, key in enumerate(col_dict.keys()):
    sel_cols = ["animal_number", "begin_time"] + col_dict[key]
    temp = all_df[sel_cols]
    # select columns and melt 
    temp2 = pd.wide_to_long(temp, [key], i = ["animal_number", "begin_time"], j="teat", sep='_', suffix=r'\w+').reset_index()
    if i == 0:
        df_long = temp2.copy()
    else:
        df_long = pd.merge(df_long, temp2, on = ["animal_number", "begin_time", "teat"])
df_long
```
