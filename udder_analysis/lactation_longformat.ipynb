```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
from datetime import date, datetime, timedelta
from sklearn import linear_model
```

```python
dirpath = os.getcwd()
data_path = os.path.join(dirpath, "..", "delpro_vms", "data")
feature_path = os.path.join(dirpath, "..", "udder_processing", "features_dict")
out_dir = "long_format_df"
# read vistit data
data_df = pd.read_csv(os.path.join(data_path, "merged_lactation.csv"))
# read features
feature_df = pd.read_csv(os.path.join(feature_path, "gmfeature_table.csv"))
feature_df = feature_df.rename(columns = lambda col: "_".join(col.split("_")[::-1]) if "_" in col else col)
```

```python
merged_long = pd.merge(feature_df, data_df, left_on = ["cow"], right_on = ["animal_number"])
merged_long.to_csv(os.path.join(out_dir, "lactation_features.csv"), index = False)
```
