```python
import pandas as pd
import os
import numpy as np
```

```python
# info
path = os.getcwd()
dirpath = os.path.join(path, "video_metadata")
filedf = pd.read_csv(os.path.join(dirpath,"video_metadata_20231117_p2.csv"))
cow_list = np.unique(filedf.cow)
filedf["time2"] = pd.to_datetime(filedf["time"],format= '%H:%M:%S')
filedf = filedf.sort_values(by='time2').reset_index(drop= True)
```

```python
filedf2 = pd.DataFrame(columns =filedf.columns)
for cow in cow_list:
    cow_files = filedf[filedf.cow == cow].reset_index(drop= True)
    filedf2 = pd.concat([filedf2, cow_files.loc[:0]],  ignore_index=True)
```

```python
with open("filelist_20231117_02.txt", "w") as f:
    f.write(",".join(filedf2.filename[:25]))
```
